We will now show you a better way to get the list of pairs of (xi, yi) instead of using for loops. This is called arrays and vectors. Unfornunately, arrays and vectors are not part of core python, but the library Numpy allows you to do create and use them.

Below we import it:

```python
import numpy as n

```

To create a range of numbers in numpy for x we use the arange() function as follows:

```python
starting_array_value = -10
ending_array_value = 10
x_array_1 = n.arange(starting_array_value, ending_array_value)
print(x_array_1)
```

Now we have easily created values of x_1. 
Now lets say we have function f() where y = f(x) = (x + 2)^2 so y = (x + 2)^2
When we want the corresponding y_1 vector of our x_1 vector we represent it as y_1 = f(x_1) = (x_1 + 2)^2
We this is how we exactly are able to do it in python using numpy. Remember we imported numpy as n.


```python
x_1 = x_array_1

y_1 = (x_1 + 2)**2 # note ** is the same as ^ in python

print(x_1)
print(y_1)

```

We can also print this side by side using a for loop

```python
for i in range(len(x_1)):
    print(x_1[i], y_1[i])
```

We will just define the function in python. 

```python
def f(x):
    y = (x+2)**3 # Here we will cube it.
    return(y) 

y_2 = f(x_1)

print(x_1)
print(y_2, '\n')

print("x_1", "y_2")
for i in range(len(x_1)):    
    print(x_1[i], y_2[i])
```

For more complicated math functions like sin(), cosine(), etc. they are contained in the numpy library and can be accessed from there to use the same way we used the arange() function from numpy: n.sin(), n.cos(), etc. Similarly, many other functions are stored in the numpy library and can be accessed this way.



```python
y_3 = n.cos(x_1)

print('x_1', 'y_3')
for i in range(len(x_1)):
    print(x_1[i], y_3[i])
```

Now we have a way to create the vector (list of numbers) of values we are interested in for x_1. 
We can define any function in python using the def command. 
And we can produce the corresponding values of the y_1. 
We can also access more complicated math functions.
All of the above in vectors using numpy extremely easily in 2 to 3 lines of code. 

We will create a few functions value pairs below and then we will begin ploting them. 



```python
def g(x):
    import numpy as n # importing the library in side the function just as a redundacy step.
    y = (n.cos(x) + n.sin(x+1) + 2) 
    return(y)

x_2 = n.arange(-15, 15)
y_4 = g(x_2)

print('x_2', 'y_4')
for i in range(len(x_2)):
    print(x_2[i], y_4[i])
```

Now we will create a function that only needs to take in the x vector and y vector and it will plot the function with nice asthetics as above.

```python
def plot_fn(x_vector,y_vector):
    
    import numpy as n
    import matplotlib.pyplot as p

    xmin = n.min(x_vector)
    xmax = n.max(x_vector)

    ymin = n.min(y_vector)
    ymax = n.max(y_vector)


    p.style.use('seaborn')
    fig, ax = plt.subplots()
    plt.axis([xmin,xmax,ymin,ymax]) # window size
    plt.plot([xmin,xmax],[0,0],'black' , linestyle = 'dotted') # black x axis
    plt.plot([0,0],[ymin,ymax], 'black', linestyle = 'dotted') # black y axis
    plt.xlabel('x')
    plt.ylabel('y')
    plt.plot(x_vector,y_vector, 'black', marker = "o") # Add points on the graph of the function to show the points that were explicitly plotted.

    plt.show()
```

After creating the function that plots we will now plot one function which is pretty complicated.

```python
x_4 = n.arange(-10, 10)

def f1(x):
    import numpy as n
    y = 2*n.cos(0.5*x+1) + n.sin(x) + x/100 #+ (x**3)/5
    return(y)

y_4 = f1(x_4)


```

```python
print('x_4', 'y_4')
for i in range(len(x_4)):
    print(x_4[i], y_4[i])

print(x_4)
print(y_4)

print('x_4 length is:', len(x_4))
print('y_4 length is:', len(y_4))
```

Now we plot

```python
plot_fn(x_4, y_4)
```

#### Drawing Constant Rate Functions and More Ways to Customize a Graph:

Below we show you how to plot slope functions as well as how to further customize your graphs when you plot them.

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = 0
xmax = 10
ymin = -10
ymax = 10
points = 2*(xmax-xmin)
x = np.linspace(xmin, xmax, points)

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'black', linestyle = 'dotted') # blue x axis
plt.plot([0,0],[ymin,ymax], 'black', linestyle = 'dotted') # blue y axis

ax.set_xlabel("x values")
ax.set_ylabel("y values")
ax.set_title("Some Graph")
ax.grid(True)

ax.set_xticks(np.arange(xmin, xmax, 1))
ax.set_yticks(np.arange(ymin, ymax, 1))

y = x
z = 2*x +1
plt.plot(x, y, label = 'y = x')
plt.plot(x,z, label='z=2x+1')
#plt.plot(x,3*x, label='steeper line')
plt.plot([4],[6], 'ro', label='point')
plt.legend()
plt.show()
```

Here above you can see two slopes. Two graphs, two functions, at start at $x = 0$. The blue line can represent how your age increases with every year passing. One thing to note is that at $x=0$ we also have $y = 0$. What thing could represent the orange line? At zero the value is not zero and every time x increases by 1, the y increases by 3. Or more generally, at zero something is a particluar non-zero value, lets call this value $b$. It could be 5, 10, 100, etc. anything, but let's say it is some value $b$. Also, every time x increases by 1 y increases by a constant value $m$. One example could be your height. At 0 years, the day you were born, your height was some value $b$ and very year that passed your height increased by $m$. This will be different for different people. What are your personal $b$ and $m$ values? 

The function for this slope/graph/plot will be $y=mx+b$. 

Here you now have an example of something in the real world represented as a function in math and how that function looks like, i.e. it's _graph_.
